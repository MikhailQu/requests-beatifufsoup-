определить какой тикер имел максимальное значение роста 01.11.2019

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np



request=requests.get('https://mfd.ru/marketdata/?id=5&group=16&mode=3&sortHeader=name&sortOrder=1&selectedDate=01.11.2019')
if request.ok:
    print('ok')

html=request.content  # сохраняем ответ в переменную html
soup = (BeautifulSoup(html))  # преобразуем ответ в 'bs4.element.ResultSet' для дальнейшего рассмотрения
soup=(soup.findAll('table'))  # находим в коде заголовок в котором содержится наша таблица. Нам нужен вророй.
table = soup[1] #сохраняем нужный фрагмент страницы
data = []
for tr in table.find_all('tr'):   # цыкл преобразует html в список(строк) состоящих из списков(столбцов). 
  data.append([td.get_text() for td in tr.find_all('td')])
data = data[2:] # избавляемся от пробела вначале

# создаем DataFrame из списка списков data
data = pd.DataFrame(data, columns = ['тикер','дата','цена','чз','ростчз%','пред закр','откр',"мин","макс",'1срвзв',"шт","руб",'сделок'])
data = data.loc[data.any(axis = 1)] # заменяем значения NA на 0
data["ростчз%"] = data["ростчз%"].apply(lambda n: -float(n[1:-1]) if n[0:1]=="−" else(float(n[1:-1]) if n[0:1]=="+" else 0) ) # преобразуем столбец отпрасывая знак "%" вконце,знаки"+" и "-" вначале, а так же преобразуем "0%" в float(0)  


data = data.sort_values("ростчз%",ascending=False)  #сортируем значения столбца по возрастанию. компания с максимальный ростом будет первой.
```
